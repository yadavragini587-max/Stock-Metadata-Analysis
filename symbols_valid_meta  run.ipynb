```python
print("SImporting required libraries...")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported successfully.\n")
```

```python
print("Step 2: Loading dataset...")

file_path = "C:/Users/Ashok/Downloads/symbols_valid_meta (1).csv"
df = pd.read_csv(file_path)

print("Dataset loaded successfully.")
print("Shape of dataset:", df.shape)
print("First 5 rows:\n")
print(df.head(), "\n")
```

```python
print("Step 3: Checking dataset information...\n")

print("Column Names:")
print(df.columns, "\n")

print("Data Types:")
print(df.dtypes, "\n")

print("Missing Values:")
print(df.isnull().sum(), "\n")
```

```python
print("Step 4: Cleaning dataset...\n")

initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
final_rows = df.shape[0]

print("Duplicates removed:", initial_rows - final_rows)

df.dropna(how='all', inplace=True)

print("Remaining shape after cleaning:", df.shape, "\n")
```

```python
print("Step 5: Generating statistical summary...\n")

print(df.describe(include='all'), "\n")
```

```python
print("Step 6: Analyzing categorical columns...\n")

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"Top 5 values in {col}:")
    print(df[col].value_counts().head(), "\n")
```

```python
print("Step 7: Creating visualization...\n")

column_to_plot = df.select_dtypes(include='object').columns[0]

top_values = df[column_to_plot].value_counts().head(10)

plt.figure()
top_values.plot(kind='bar')
plt.title(f"Top 10 {column_to_plot}")
plt.xlabel(column_to_plot)
plt.ylabel("Count")
plt.xticks(rotation=45)

print("Displaying plot...")
plt.show()
```

```python
print("Step 8: Saving cleaned dataset...\n")

df.to_csv("cleaned_symbols_data.csv", index=False)

print("File saved as cleaned_symbols_data.csv")
```
